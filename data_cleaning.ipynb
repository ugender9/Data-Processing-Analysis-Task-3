```python
import pandas as pd

# Create sample dataset and save to CSV
data = {
    'brand': [' Toyota', 'ford ', 'Toyota', 'BMW', 'BMW', ' Ford', 'Toyota', None],
    'mileage': ['50000 km', '30000 miles', '45000', '100k', '20000', '15000', '45000', '10k'],
    'selling_price': [15000, 12000, 14500, 25000, 26000, 11000, 14500, None],
    'year': [2015, 2018, 2015, 2020, 2021, 2017, 2015, 2019]
}

pd.DataFrame(data).to_csv('used_cars.csv', index=False)
print('used_cars.csv created successfully!')
```

# ✅ Task 1 — Explore and Identify Issues

```python
import pandas as pd

# Load dataset
df = pd.read_csv("used_cars.csv")

# Basic exploration
print("Shape:", df.shape)
print("\nInfo:")
display(df.info())

print("\nDescribe:")
display(df.describe())

print("\nMissing values:")
print(df.isnull().sum())
```

### 🔍 Issues Identified:
- `selling_price` contains null values.
- `mileage` column is stored as an object/string instead of numeric.
- `brand` column has inconsistent formatting (leading spaces/mixed case).
- Duplicate rows are present.

# ✅ Task 2 — Clean the Data

```python
# 1. Drop rows where target (selling_price) is null
df = df.dropna(subset=['selling_price'])

# 2. Handle missing values (brand mode imputation)
df['brand'] = df['brand'].fillna(df['brand'].mode()[0])

# 3. Clean 'brand' column
df['brand'] = df['brand'].str.strip().str.lower()

# 4. Extract numeric values from mileage
df['mileage'] = df['mileage'].astype(str).str.extract(r'(\d+)').astype(float)

# 5. Remove duplicates
df = df.drop_duplicates()

# Final check
print(df.info())
display(df.head())
```

# ✅ Task 3 — Compute Baseline MAE

```python
from sklearn.metrics import mean_absolute_error

# Baseline model = predict mean selling price
mean_price = df['selling_price'].mean()

# Predictions (same value for all rows)
predictions = [mean_price] * len(df)
actual = df['selling_price']

# Compute MAE
mae = mean_absolute_error(actual, predictions)

print(f"Baseline Mean Selling Price: {mean_price:.2f}")
print(f"Baseline MAE: {mae:.2f}")
```
