## Task 1 — Clean the Data
In this step, we initialize the dataset, impute missing values for 'age' (median) and 'city' (mode), and remove any duplicate records.

```python
import pandas as pd
import numpy as np

# Create dataset
df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age':         [25, np.nan, 35, np.nan, np.nan, 40],
    'city':        ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender':      ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})

print("Original Data:")
display(df)

# 1. Impute missing values
df['age'] = df['age'].fillna(df['age'].median())
df['city'] = df['city'].fillna(df['city'].mode()[0])

# 2. Remove duplicate rows
df = df.drop_duplicates()

print("\nCleaned Data:")
display(df)
```

## Task 2 — Encode Categorical Columns
We use One-Hot Encoding for the 'city' column to avoid ordinal bias and Label Encoding for 'gender'.

```python
from sklearn.preprocessing import LabelEncoder

# One-Hot Encoding for city
df_encoded = pd.get_dummies(df, columns=['city'])

# Label Encoding for gender
le = LabelEncoder()
df_encoded['gender'] = le.fit_transform(df_encoded['gender'])

print("Encoded Data:")
display(df_encoded.head())
```

## Task 3 — Scale Numerical Features
We demonstrate two common scaling techniques: Min-Max Scaling (normalization) and Standardization.

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Columns to scale
num_cols = ['age', 'annual_income']

# ---- Min-Max Scaling ----
minmax_scaler = MinMaxScaler()
df_minmax = df_encoded.copy()
df_minmax[num_cols] = minmax_scaler.fit_transform(df_minmax[num_cols])

print("Min-Max Scaled Data:")
display(df_minmax.head())

# ---- Standardization ----
std_scaler = StandardScaler()
df_standard = df_encoded.copy()
df_standard[num_cols] = std_scaler.fit_transform(df_standard[num_cols])

print("\nStandardized Data:")
display(df_standard.head())
```

## 🧠 Explanation of Scaling Techniques

👉 **Min-Max Scaling**
* Scales values between **0 and 1**.
* Best when data has **no extreme outliers** and you need bounded values (e.g., neural networks).

👉 **Standardization**
* Transforms data to **mean = 0, std = 1**.
* Preferred when data has **outliers** or when using models like Logistic Regression or SVM.
